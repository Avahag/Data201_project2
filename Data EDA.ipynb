```python
#convert categorical data to numeric for heatmap 
data_encoded=pd.get_dummies(data)
data_encoded
```

```python
# checking for missing values 
missing_count = data.isnull().sum()
print(missing_count)
```

```python
# data is clean (no missing values) 
```

```python
# Visualize distributions
sns.boxplot(data['Hours_Worked_Per_Week'])
plt.show()
```

```python
# Distribution of Productivity Score
sns.histplot(data['Productivity_Score'], kde=True)
plt.title('Distribution of Productivity Score')
plt.show()
```

```python
correlation_matrix = data_encoded.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Employment_Type', y='Hours_Worked_Per_Week', data=data, palette='viridis')  
plt.title('Hours Worked Per Week by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Hours Worked Per Week')
plt.show()
```

```python
# Group analysis by employment type
plt.figure(figsize=(8, 6))
sns.boxplot(x='Employment_Type', y='Productivity_Score', data=data , palette='coolwarm')
plt.title('Productivity Score by Employment Type')
plt.show()
```
